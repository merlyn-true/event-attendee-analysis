# **Import Packages**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # set seaborn chart style as the default
```

# **Import Pre-processed Data**

```python
# Set folder path
folder_path = r"..\data\preprocessed"

# Zoom session data
zoom_session_data = pd.read_csv(folder_path+r'\zoom-session-data.csv')

# Attendee location data
attendee_location_data = pd.read_csv(folder_path+r'\attendee-location-data.csv')

# International attendee data
international_attendee_data = pd.read_csv(folder_path+r'\international-attendee-data.csv')
international_attendee_data.rename(columns={'please_specify': 'location'}, inplace=True)
# Remove na from DataFrame
mask = international_attendee_data['location'] == 'na'
international_attendee_data = international_attendee_data[~mask].copy().reset_index(drop=True)
```

# **Visualizations**
To visualize the data I will create bar charts for the location data and histograms for the zoom session data

## **Bar Charts**

```python
attendee_location_data.head()
```

```python
# Set seaborn style to white
sns.set_style('white')
# Create a 9x6 figure
plt.figure(figsize=(9,6))
# Plot bar chart
plt.bar(x=attendee_location_data['province_territory'],
        height=attendee_location_data['count'],
        color='midnightblue')
# Add title
plt.title('Attendees by Location', fontsize=16, fontweight='bold')
# Add y-axis label, set fontsize to 13
plt.ylabel('Attendee Count', fontsize=13)
# Set y tick marks to 13
plt.yticks(fontsize=13)
# Rotate x axis labels to 45 degrees, set fontsize to 13
plt.xticks(rotation=75, fontsize=13)
# Remove top and right borders
sns.despine()
# Show plot
plt.show()
```
