# **1. Setup & Imports**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set seaborn chart style as the default
sns.set() 
```

# **2. Load Preprocessed Data**

```python
# Set folder path
folder_path = r"..\data\preprocessed"

# Zoom session data
zoom_session_data = pd.read_csv(folder_path+r'\zoom-session-data.csv')

# Attendee location data
attendee_location_data = pd.read_csv(folder_path+r'\attendee-location-data.csv')

# International attendee data
international_attendee_data = pd.read_csv(folder_path+r'\international-attendee-data.csv')
international_attendee_data.rename(columns={'please_specify': 'location'}, inplace=True)

# Remove na from DataFrame
mask = international_attendee_data['location'] == 'na'
international_attendee_data = international_attendee_data[~mask].copy().reset_index(drop=True)
```

# **3. Create Visualizations**
To visualize the data I will create bar charts for the location data and histograms for the zoom session data

## **3.1. Bar Charts**

```python
attendee_location_data.head()
```

```python
# Set seaborn style to white
sns.set_style('white')
# Create a 9x6 figure
fig, ax = plt.subplots(figsize=(9,6))
# Plot bar chart
bars = ax.bar(x=attendee_location_data['province_territory'],
    height=attendee_location_data['count'],
    color='midnightblue')
# Add title
ax.set_title('Attendees by Location', fontsize=16, fontweight='bold')
# Add y-axis label, set fontsize to 13
ax.set_ylabel('Attendee Count', fontsize=13)
# Set y tick marks to 13
ax.tick_params(axis='y', labelsize=13)
# Rotate x axis labels to 75 degrees, set fontsize to 13
plt.xticks(rotation=75, fontsize=13)
# Remove top and right borders
sns.despine()
# Add labels on top of bars
ax.bar_label(bars, fmt='%d')
# Save plot
plt.savefig('../output/charts-figures/attendees-by-location.png')
# Show plot
plt.show()
```

## **3.2. Histograms**

```python
zoom_session_data.head()
```

### **3.2.1. Time in Session**

```python
# Create 8x6 figure
plt.figure(figsize=(8,6))
# Create histogram
ax = sns.histplot(data=zoom_session_data, 
             x="time_in_session", 
             binwidth=10, 
            #  edgecolor='black', 
             color='midnightblue',
             alpha=1)
# Add title
plt.title('Distribution of Time in Session', fontsize=16, weight='bold')
# Add x label
plt.xlabel("Time in Session", fontsize=13)
# Add ylabel
plt.ylabel("Attendee Coount", fontsize=13)
# Remove top and right border
sns.despine()
# Add labels
for container in ax.containers:
    ax.bar_label(container, fmt='%d')  # fmt controls label format
# Add average time in session
avg_time_in_session = zoom_session_data['time_in_session'].mean()
ax.axvline(x=avg_time_in_session, 
           color="red", 
           linestyle="--", 
           linewidth=2, 
           label=f'Average time in session: {int(np.ceil(avg_time_in_session))} minutes')
# Show legend
ax.legend()
# Save plot
plt.savefig('../output/charts-figures/time-in-session-histogram.png')
# Show plot
plt.show()
```

### **3.2.2. Join Time**

```python
# Create 8x6 figure
plt.figure(figsize=(8,6))
# Create histogram
ax = sns.histplot(data=zoom_session_data, 
             x="join_minute", 
             binwidth=10, 
            #  edgecolor='black', 
             color='midnightblue',
             alpha=1)
# Add title
plt.title('Distribution of Join Time', fontsize=16, weight='bold')
# Add x label
plt.xlabel("Join Time", fontsize=13)
# Add ylabel
plt.ylabel("Attendee Coount", fontsize=13)
# Remove top and right border
sns.despine()
# Add labels
for container in ax.containers:
    ax.bar_label(container, fmt='%d')  # fmt controls label format
# Add average join time
avg_join_time = zoom_session_data['join_minute'].mean()
ax.axvline(x=avg_join_time, 
           color="red", 
           linestyle="--", 
           linewidth=2, 
           label=f'Average join time: {int(np.ceil(avg_join_time))} minutes')
# Show legend
ax.legend()
# Save plot
plt.savefig('../output/charts-figures/join-time-histogram.png')
# Show plot
plt.show()
```

### **3.2.3 Leave Time**

```python
# Create 8x6 figure
plt.figure(figsize=(8,6))
# Create histogram
ax = sns.histplot(data=zoom_session_data, 
             x="leave_minute", 
             binwidth=10, 
            #  edgecolor='black', 
             color='midnightblue',
             alpha=1)
# Add title
plt.title('Distribution of Leave Time', fontsize=16, weight='bold')
# Add x label
plt.xlabel("Leave Time", fontsize=13)
# Add ylabel
plt.ylabel("Attendee Coount", fontsize=13)
# Remove top and right border
sns.despine()
# Add labels
for container in ax.containers:
    ax.bar_label(container, fmt='%d')  # fmt controls label format
# Add average leave time
avg_leave_time = zoom_session_data['leave_minute'].mean()
ax.axvline(x=avg_leave_time, 
           color="red", 
           linestyle="--", 
           linewidth=2, 
           label=f'Average leave time: {int(np.ceil(avg_leave_time))} minutes')
# Show legend
ax.legend()
# Save plot
plt.savefig('../output/charts-figures/leave-time-histogram.png')
# Show plot
plt.show()
```

- **125 participants** remained in the session for **over 150 minutes (2.5 hours)**.

## **4.3. Join Time Behavior**
- **243 participants** joined the session within the **first 20 minutes**.
- On average, participants joined **34 minutes after** the session started.

## **4.4. Leave Time Behavior**
- **300 participants** stayed past the **150-minute (2.5-hour)** mark before leaving.
- The **average leave time** was **140 minutes**, which is approximately **2 hours and 20 minutes** into the session.
